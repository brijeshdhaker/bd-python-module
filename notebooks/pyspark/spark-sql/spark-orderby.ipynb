Clarification with Spark SQL
The confusion between "sort" and "orderBy" often stems from the different behavior of the corresponding clauses in Spark SQL: 

|**Feature**| df.sort() / df.orderBy()  (DataFrame API) |SORT BY (Spark SQL clause)   | ORDER BY (Spark SQL clause) |
|------|------|------|-----   |
|**Sorting Scope**|   Global total order across the entire DataFrame  | Partition-wise order (sorts data within each partition).|Global total order across the entire result set.   |
|**Performance**| Slower due to the global data shuffle| Faster as it avoids a full shuffle.| Slower due to the global data shuffle.|


```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Create Spark session
spark = SparkSession.builder.appName("example").getOrCreate()

# Sample data
data = [("James", "Smith", "M", 30),
        ("Anna", "Rose", "F", 41),
        ("Robert", "Williams", "M", 62)]

# Create DataFrame
columns = ["firstname", "lastname", "gender", "age"]
df = spark.createDataFrame(data, columns)



```

```python
# Order by age ascending
df_ordered = df.orderBy("age")
df_ordered.show()
```

```python
# Order by age descending
df_ordered_desc = df.orderBy(col("age").desc())
df_ordered_desc.show()
```

```python
# Sort by age ascending
df_sorted = df.sort("age")
df_sorted.show()

```

```python
# Sort by age descending
df_sorted_desc = df.sort(col("age").desc())
df_sorted_desc.show()
```
