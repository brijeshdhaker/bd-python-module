
### Question - 008: - Write a pyspark query using below input data to get provided output

Input Data : 
|  C1|    C2|      C3|
|----|------|--------|
| Goa|      |  Mumbai|
|    |Mumbai|    NULL|
|NULL|      |Banglore|


Output Data :

|  Output|
|--------|
|     Goa|
|  Mumbai|
|Banglore|

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession
    .builder
    .appName("spark-q-008")
    .master("local[*]")
    .config("spark.ui.port", "4040")
    .getOrCreate()
)

#
print(spark.sparkContext.uiWebUrl)

#
spark.sparkContext.setLogLevel("ERROR")

#
# 
#
def display(df):
    df.show(truncate=False)
```

```python
# Sample data
data = [
    ("Goa", "", "Mumbai"),
    ("", "Mumbai", None),
    (None, "", "Banglore")
]

# Create a DataFrame
columns = ["C1", "C2", "C3"]
df = spark.createDataFrame(data, columns)

# Show the initial DataFrame
df.show()
```

```python
from pyspark.sql.functions import *

df.withColumn("C4", coalesce("C1","C2","C3")).show()
```

```python

df.withColumn("C4", coalesce(
        when(col("C1")=="", None).otherwise(col("C1")),
        when(col("C2")=="", None).otherwise(col("C2")),
        when(col("C3")=="", None).otherwise(col("C3"))
    )
).select(col("C4").alias("Output")).show()
```
