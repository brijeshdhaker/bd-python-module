##### Break down two essential transformations: Flatten and Explode in PySpark!

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession
    .builder
    .appName("spark-q-005")
    .master("local[*]")
    .config("spark.ui.port", "4040")
    .getOrCreate()
)

#
print(spark.sparkContext.uiWebUrl)

#
spark

#
# 
#
def display(df):
    df.show(truncate=False)
```

```python

#
data = [
    (1, 'Alice', ['Python', 'SQL']),
    (2, 'Bob', ['Java', 'Python', 'Scala','SQL']),
    (3, 'Charlie', ['SQL', 'R']),
    (4, 'David', ['Python', 'Java', 'Pyspark']),
    (5, 'Eve', ['SQL', 'Databricks'])
]

#
schema = ['id', 'name', 'skills']

#
df = spark.createDataFrame(data, schema)
display(df)

```

```python
from pyspark.sql.functions import *

df.select(col("id"),col("name"),explode(col("skills")).alias("skill")).show()
```
