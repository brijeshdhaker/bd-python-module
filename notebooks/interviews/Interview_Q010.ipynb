### Question - 010: - Write a pyspark query using below input data to get provided output

Input Data : 

|     Names|Country|
|----------|-------|
|John,Smith| Canada|
|Mike,David|    USA|

Output : 

|Country| Name|
|-------|-----|
| Canada| John|
| Canada|Smith|
|    USA| Mike|
|    USA|David|


```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("spark-q-0010").getOrCreate()
```

```python

data = [("John,Smith", "Canada"), ("Mike,David", "USA")]
df = spark.createDataFrame(data,["Names","Country"])
df.show()
```

```python
from pyspark.sql.functions import *

df.withColumn("Name", explode(split(col("Names"), ","))).drop(col("Names")).show()
```
